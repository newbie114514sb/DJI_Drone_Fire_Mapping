```python
%pip install roboflow python-dotenv

import os
from getpass import getpass
from pathlib import Path

from dotenv import load_dotenv
from roboflow import Roboflow

for env_path in (Path.cwd() / ".env", Path.cwd().parent / ".env"):
    if env_path.exists():
        load_dotenv(env_path, override=False)

api_key = os.environ.get("ROBOFLOW_API_KEY", "").strip()
if not api_key:
    api_key = getpass("Enter ROBOFLOW_API_KEY: ").strip()
    if api_key:
        os.environ["ROBOFLOW_API_KEY"] = api_key

if not api_key:
    raise ValueError("ROBOFLOW_API_KEY was not found. Add it to a .env file or paste it when prompted.")

rf = Roboflow(api_key=api_key)
project = rf.workspace("new-f7fvn").project("visdrone-ozubn-mfdts")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
%pip install --upgrade ultralytics
%pip install --upgrade torch torchvision --index-url https://download.pytorch.org/whl/cu128

import torch
print("torch version:", torch.__version__)
print("cuda available:", torch.cuda.is_available())
print("cuda version:", torch.version.cuda)
if torch.cuda.is_available():
    print("device 0:", torch.cuda.get_device_name(0))
```

```python
%pip install ultralytics
import torch
from ultralytics import YOLO

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

model = YOLO("yolov8n.pt")

results = model.train(
    data=dataset.location + "/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16 if device != "cpu" else 8,  # larger batch on GPU
    device=device,
    name="visdrone_firetrain",
    project="runs/detect",
    save_period=10,  # save checkpoint every 10 epochs
)

print("Training complete. Best weights saved to:", results.save_dir)
```
