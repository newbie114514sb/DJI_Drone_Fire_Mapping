```python
%pip install roboflow

import os
from roboflow import Roboflow

api_key = os.environ.get("ROBOFLOW_API_KEY", "")
if not api_key:
    raise ValueError("Set the ROBOFLOW_API_KEY environment variable before running.")

rf = Roboflow(api_key=api_key)
project = rf.workspace("new-f7fvn").project("visdrone-ozubn-mfdts")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
%pip install ultralytics
import torch
from ultralytics import YOLO

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

model = YOLO("yolov8n.pt")

results = model.train(
    data=dataset.location + "/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16 if device != "cpu" else 8,  # larger batch on GPU
    device=device,
    name="visdrone_firetrain",
    project="runs/detect",
    save_period=10,  # save checkpoint every 10 epochs
)

print("Training complete. Best weights saved to:", results.save_dir)
```
